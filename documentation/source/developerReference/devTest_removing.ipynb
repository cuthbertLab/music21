```python
oct1 = [0, 1, 3, 4, 6, 7, 9, 10]
oct2 = [(n+1) % 12 for n in oct1]
oct3 = [(n+2) % 12 for n in oct1]
oct1
```

```python
lastPitchClass = -1

for n in sacrem.recurse().notes:
    n.pitch.simplifyEnharmonic(inPlace=True)
    pc = n.pitch.pitchClass
    n.lyric = ''
    if pc == lastPitchClass:
        n.style.hideObjectOnPrint = True
        n.stemDirection = 'noStem'
        continue

    lastPitchClass = pc
        
    if pc in oct1:
        n.style.color = 'red'
        n.lyric += '1'
    if pc in oct2:
        n.notehead = 'x'
        n.lyric += '2'

    if pc in oct3:
        n.stemDirection = 'noStem'
        n.lyric += '3'
    
```
